```python
geolocalizador = Nominatim(user_agent="http-p")
```

```python
geolocalizador.geocode([colombia.iloc[0]['ADM1NAME'],colombia.iloc[0]['ADM2NAME']])
```

```python
def localizar(df):
    localizaciones = []
    errores = []
    for i in range(len(df)):
        l = geolocalizador.geocode([df.iloc[i]['ADM1NAME'],df.iloc[i]['ADM2NAME']])
        try:
            localizaciones.append({    
                'estado' :  df.iloc[i]['ADM1NAME'], 
                'ciudad' : df.iloc[i]['ADM2NAME'],
                'latitud': l[-1][0],
                'longitud': l[-1][1],
                'detalles': l
            })
        except TypeError:
            errores.append({
                'estado' :  df.iloc[i]['ADM1NAME'], 
                'ciudad' : df.iloc[i]['ADM2NAME'],               
                'detalles': l
            })
        print(f"Realizado el {(round(i/len(df)))*100}%")

    loca = pd.DataFrame(localizaciones)
    data_geo = gpd.GeoDataFrame(loca, geometry = gpd.points_from_xy(loca.longitud,loca.latitud ))
    return data_geo, pd.DataFrame(errores)
```
